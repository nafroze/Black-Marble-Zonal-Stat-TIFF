```python
import geopandas as gpd
import rasterio
from rasterio.plot import show
from rasterio.plot import show_hist
import rasterstats
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
os.chdir(r'C:\Users\Nazia\research_ntl\Mosaiced data\tiff_feb')
rasterFiles = os.listdir(os.getcwd())
len(rasterFiles)
```

```python
shapefile = gpd.read_file(r"C:\Users\Nazia\research_ntl\mask\houston_shapefile.shp")
```

```python
median_nlr = []
```

```python
for i in range(len(rasterFiles)):
    # Read the nighttime raster
    rf = rasterio.open(rasterFiles[i], mode = 'r')
    
    # Assign raster values to a numpy nd array
    nighttime_array = rf.read(1)
    
    affine = rf.transform
    
    # Calculating the zonal statistics 
    med_rf  = rasterstats.zonal_stats(shapefile, nighttime_array, affine = affine,
                                      stats = ['median'], 
                                      geojson_out = True)
    
    median_nlr.append(med_rf[0]['properties'])
```

```python
len(median_nlr)
```

```python
median_ntL_houston = pd.DataFrame(median_nlr)
```
